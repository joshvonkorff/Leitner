```python
import re

pattern = r"[a-zA-Z_][a-zA-Z0-9_]+\("

filename = "python.txt"
num = 0

d = dict()
with open(filename, "r") as file:
    for line in file:
        result = re.findall(pattern, line)
        if result:
            d[result[0]] = d.setdefault(result[0], 0) + 1
        num += 1
        if num % 1000000 == 0:
            d = {k: d[k] for k in d if d[k] > 5}
        if num >= 100000000:
            break
            
print("Number of lines processed: ", num)
            
print(sorted(d.items(), key = lambda x: x[1], reverse = True))
```
